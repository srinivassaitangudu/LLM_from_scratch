```python
import re
result = re.split(r'(\s)', text)
```

```python
result
```

```python
result = [item for item in result if item.strip()]
preprocessed_text = result
```

```python
# converting tokens into token ids
```

```python
all_words = sorted(set(preprocessed_text))
vocab_size = len(all_words)
print(f"Vocab size: {vocab_size}")
```

```python
vocab = {token:integer for integer, token in enumerate(all_words)}
```

```python
print(vocab)
```

```python
class SimpleTokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {id_: token for token, id_ in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'(\s)', text)
        preprocessed = [token for token in preprocessed if token.strip()]
        ids = [self.str_to_int[token] for token in preprocessed]
        return ids
        

    def decode(self, ids):
        text = ''.join([self.int_to_str[id_] for id_ in ids])
        text = re.sub(r'\s+', ' ', text)
        return text
        
```

```python
tokeniser = SimpleTokenizer(vocab)
```

```python
text = "its the last time he painted, you know."
```

```python
ids = tokeniser.encode(text)
```

```python
ids
```

```python
tokeniser.decode(ids)
```

```python
all_tokens  = sorted(list(set(preprocessed_text)))
all_tokens.extend(["<|endoftext|>", "<|unknown|>"])

vocab = {token:integer for integer, token in enumerate(all_tokens)}
```

```python
len(vocab.items())
```

```python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {id_: token for token, id_ in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'(\s)', text)
        preprocessed = [token for token in preprocessed if token.strip()]
        preprocessed = [token if token in self.str_to_int else "<|unknown|>" for token in preprocessed]
        ids = [self.str_to_int[token] for token in preprocessed]
        return ids
        

    def decode(self, ids):
        text = ''.join([self.int_to_str[id_] for id_ in ids])
        text = re.sub(r'\s+', ' ', text)
        return text
```

```python
tokeniser = SimpleTokenizerV2(vocab)
```

```python
tokeniser.encode(text)
```

```python
tokeniser.decode(tokeniser.encode(text))
```
